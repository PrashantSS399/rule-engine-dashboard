```python
from sqlalchemy import create_engine
import pandas as pd

# Replace these with your actual credentials
DB_CONFIG = {
    'host': 'localhost',
    'database': 'Data_Asset_Linkage',
    'user': 'postgres',
    'password': 'password',
    'port': '5432'           
}

# Create connection string
conn_string = f"postgresql+psycopg2://{DB_CONFIG['user']}:{DB_CONFIG['password']}@{DB_CONFIG['host']}:{DB_CONFIG['port']}/{DB_CONFIG['database']}"

# Test connection
try:
    engine = create_engine(conn_string)
    with engine.connect() as conn:
        print("Connection successful!")
        # Example query
        df = pd.read_sql("SELECT version()", conn)
        print(df.iloc[0,0])
except Exception as e:
    print(f"Connection failed: {e}")
```

```python
customers = pd.read_sql_table('customer', engine)
```
